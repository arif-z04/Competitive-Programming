```python
import numpy as np

# Define the matrix
A = np.array([[3, 4, -1],
              [1, 0, 3],
              [2, 5, -4]])

```

```python
# 1. Determinant
det_A = np.linalg.det(A)
print(f"Determinant of A: {det_A:.0f}")

# 2. Adjoint = Transpose of cofactor matrix
def adjoint(matrix):
    cofactors = np.zeros(matrix.shape)
    for row in range(matrix.shape[0]):
        for col in range(matrix.shape[1]):
            # Minor matrix
            minor = np.delete(np.delete(matrix, row, axis=0), col, axis=1)
            # Cofactor with sign
            cofactors[row, col] = ((-1) ** (row + col)) * np.linalg.det(minor)
    return cofactors.T  # Transpose to get adjoint

adj_A = adjoint(A)
print("\nAdjoint of A:")
print(adj_A)

# 3. Inverse (if determinant is not zero)
if det_A != 0:
    inv_A = np.linalg.inv(A)
    print("\nInverse of A:")
    print(inv_A)
else:
    print("\nMatrix is singular, no inverse exists.")
```

```python
rank_A = np.linalg.matrix_rank(A)
print(f"Rank: {rank_A}")
```
